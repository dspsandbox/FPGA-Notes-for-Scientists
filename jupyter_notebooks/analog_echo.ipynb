```python
import pynq
ol = pynq.Overlay("analog_echo.bit")
```

```python
offset_1 = -0.50 #Volts
offset_2 = 0.25 #Volts

offset_1 = int(offset_1 * (2**15 - 1))
offset_2 = int(offset_2 * (2**15 - 1))

ol.axi_gpio_0.channel1.write(val=offset_1, mask=0xffff)
ol.axi_gpio_0.channel2.write(val=offset_2, mask=0xffff)
```
